```python
import pyspark
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder.appName("MyApp").getOrCreate()
df = spark.read.csv("/Volumes/workspace/default/olist/C1.csv",header = True,inferSchema=True)
df.show()
df.printSchema()
```

```python
df.display()
```

```python
df = df.replace(["blank","Blank"],None)
df = df.replace(["india","India"],"IN")
df = df.replace("New York", "USA")
df.display()
```

```python
df = df.na.fill({
    "Category" : "Unknown",
    "Sales" : 0,
    "JoinDate" : "01-01-2023"
})
df.display()
```

```python
df = df.dropDuplicates(["CustomerID"])
df.display()
```
