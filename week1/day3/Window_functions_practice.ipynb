```sql
-- 1.	Assign a unique row number to all employees based on salary (highest first). 
select *, row_number() over(order by salary desc) as row_num from employees
```

```sql
-- 2.	Assign row numbers to employees within each department based on salary descending. 
select *, row_number() over(partition by department order by salary desc) as row_num from employees
```

```sql
-- 3.	Assign row numbers based on employee joining date (latest first). 
select *, row_number() over(order by join_date desc) as row_num from employees
```

```sql
-- 4.	Assign row numbers within each department based on earliest joining date. 
select *, row_number() over(partition by department order by join_date asc) as row_num from employees
```

```sql
-- 5.	Assign row numbers to employees based on salary (lowest first). 
select *, row_number() over(order by salary asc) as row_num from employees
```

```sql
-- 6.	Assign row numbers within department for employees based on name alphabetically.
select *, row_number() over(order by emp_name asc) as row_num from employees
```

```sql
-- 7.	Rank all employees based on salary (highest first). 
select *, rank() over (order by salary desc) as rank from employees
```

```sql
-- 8.	Rank employees within each department based on salary. 
select *, rank() over(partition by department order by salary desc) as rank from employees
```

```sql
-- 9.	Rank employees based on joining date (latest gets rank 1). 
select *, rank() over(order by join_date desc) as rank from employees
```

```sql
-- 10.	Rank employees within department based on salary (lowest first). 
select *, rank() over(partition by department order by salary asc) as rank from employees
```

```sql
-- 11.	Rank employees based on name alphabetically. 
select *, rank() over(order by emp_name asc) as rank from employees
```

```sql
-- 12.	Assign dense rank to employees based on salary (highest first). 
select *, dense_rank() over(order by salary desc) as dense_rank from employees
```

```sql
-- 13.	Assign dense rank within each department based on salary.
select *, dense_rank() over(partition by department order by salary desc) as dense_rank from employees
```

```sql
-- 14.	Assign dense rank to employees based on joining date.
select *, dense_rank() over(order by join_date) as dense_rank from employees 
```

```sql
-- 15.	Assign dense rank to employees based on salary (lowest first).
select *, dense_rank() over(order by salary asc) as dense_rank from employees 
```

```sql
-- 16.	Assign dense rank within department based on joining date. 
select *, dense_rank() over(partition by department order by join_date desc) as dense_rank from employees
```

```sql
-- TASK 1: Top 3 salaried employees per department (Window Function – RANK)
select * from (select *, dense_rank() over(partition by department order by salary desc) as rank_for_employess from employees) t where rank_for_employess <= 3
```
