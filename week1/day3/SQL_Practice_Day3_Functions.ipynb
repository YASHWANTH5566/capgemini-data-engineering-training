```sql
-- 1. Show all employees whose salary is NULL

select * from Employees1 where salary is null;
```

```sql
-- 2. Show all orders where discount is NOT NULL

select * from Orders where discount is not null;
```

```sql
-- 3. Get products where category is NULL

select * from Products where category is null;
```

```sql
-- 4. Count number of employees with NULL manager_id
select count(*) from Employees1 where manager_id is null;
```

```sql
-- Replace NULL salary with 0
select *, IFNULL(salary,0) as salary from Employees1
```

```sql
-- Replace NULL bonus with 1000
select *, IFNULL(bonus,1000) as bonus from Employees1
```

```sql
-- Show order amount, if NULL replace with 500
select *, IFNULL(amount,500) as amount from Orders
```

```sql
-- Replace NULL stock with 0
select *, ifnull(stock,0) as stock from Products
```

```sql
-- Show employee earnings using: salary, if NULL use bonus
select emp_id, name, coalesce(salary,bonus) as earnings from Employees1
```

```sql
-- Show first available value: salary → bonus → 0
select *, coalesce(salary,bonus,0) as earnings from Employees1
```

```sql
-- Show product price: price → 1000 (default)
select *, coalesce(price,1000) as price from Products
```

```sql
-- Get customer payment: amount → discount → 0
select *, coalesce(amount,discount,0) as payment from Orders
```

```sql
-- Convert salary to NULL if salary = 0
SELECT NULLIF(salary, 0) AS salary FROM Employees1;
```

```sql
-- Convert discount to NULL if discount = 0
SELECT NULLIF(discount, 0) AS discount FROM Orders;
```

```sql
SELECT amount / NULLIF(discount, 0) AS final_value
FROM Orders;
```

```sql
SELECT NULLIF(coupon_code, 'DISC10') AS coupon_code
FROM Orders;
```

```sql
-- 1️. Calculate total earnings (salary + bonus, handle NULL properly)
-- Calculate total earnings
SELECT 
    emp_id,
    COALESCE(salary, 0) + COALESCE(bonus, 0) AS total_earnings
FROM Employees1;
```

```sql
-- 2️. Show employees where both salary AND bonus are NULL
-- Employees with both salary and bonus NULL
SELECT *
FROM Employees1
WHERE salary IS NULL 
  AND bonus IS NULL;
```

```sql
-- 3️. Show products where price is NULL but category is NOT NULL
-- Products with NULL price but valid category
SELECT *
FROM Products
WHERE price IS NULL 
  AND category IS NOT NULL;
```

```sql
-- 4️. Show orders where both amount and discount are NULL
-- Orders with both amount and discount NULL
SELECT *
FROM Orders
WHERE amount IS NULL 
  AND discount IS NULL;
```

```sql
-- 5️. Show employee income using fallback logic
-- Employee income using COALESCE
SELECT 
    emp_id,
    COALESCE(salary, bonus, 1000) AS income
FROM Employees1;
```

```sql
-- 6️. Replace empty discount with NULL using NULLIF
-- Convert 0 discount to NULL
SELECT 
    order_id,
    NULLIF(discount, 0) AS discount
FROM Orders;
```

```sql
-- 7️. Show final payable amount (amount - discount, handle NULL)
-- Final payable amount
SELECT 
    order_id,
    COALESCE(amount, 0) - COALESCE(discount, 0) AS final_amount
FROM Orders;
```

```sql
-- 8️. Find employees where salary is NULL but manager exists
-- Employees with NULL salary but having a manager
SELECT *
FROM Employees1
WHERE salary IS NULL 
  AND manager_id IS NOT NULL;
```
