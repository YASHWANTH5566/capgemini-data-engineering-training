### Real World DataSet Practice

```sql
%python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.appName("Practice Real Dataset").getOrCreate()
df = spark.read.csv('/Volumes/workspace/default/olist/Practice_Real_Dataset/Capgemini Databricks VITB-2026 (1).csv', header = True, inferSchema = True)
df.show()
```

```sql
%python
import re

# Clean column names
new_cols = [re.sub(r'[^a-zA-Z0-9]', '_', c).lower() for c in df.columns]
df = df.toDF(*new_cols)
```

```sql
%python
df.show()
```

```sql
%python
from pyspark.sql.functions import col, count, when

# Count non-null values for each column
non_null_counts = df.select([count(when(col(c).isNotNull(), c)).alias(c) for c in df.columns])
non_null_counts.show()
```

```sql
%python
# Keep only columns with at least one non-null value
columns_to_keep = [c for c in df.columns if df.select(c).na.drop().count() > 0]

df_clean = df.select(columns_to_keep)
```

```sql
%python
df_clean.display()
```

```sql
%python
df_clean.write.format("delta").mode("overwrite").saveAsTable("training_enrolled")
```

```sql
select * from training_enrolled limit 10;
```
