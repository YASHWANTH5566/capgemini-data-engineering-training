```sql
-- Base Tables (Normalize Emails)
WITH student_base AS (
    SELECT 
        regno AS student_id,
        LOWER(COALESCE(college_domain_mailid, personal_mail_id)) AS email
    FROM training_enrolled
),

responses AS (
    SELECT 
        LOWER(email_address) AS email,
        timestamp
    FROM task_1_response_2
)

SELECT * FROM student_base;
```

```sql
-- Map Emails to Student IDs
WITH responses AS (
    SELECT 
        LOWER(email_address) AS email,
        timestamp
    FROM task_1_response_2
)

SELECT 
    m.student_id,
    r.email,
    r.timestamp
FROM responses r
LEFT JOIN student_email_map m
    ON LOWER(r.email) = LOWER(m.email);
```

```sql
-- Detect Duplicates (ROW_NUMBER)
WITH responses AS (
    SELECT 
        LOWER(email_address) AS email,
        timestamp
    FROM task_1_response_2
),
mapped_responses AS (
    SELECT 
        m.student_id,
        r.email,
        r.timestamp
    FROM responses r
    LEFT JOIN student_email_map m
        ON LOWER(r.email) = LOWER(m.email)
)

SELECT 
    student_id,
    email,
    timestamp,
    ROW_NUMBER() OVER (
        PARTITION BY student_id 
        ORDER BY timestamp
    ) AS rn
FROM mapped_responses;
```

```sql
-- Classification (Submitted / Duplicate / Invalid)
WITH responses AS (
    SELECT 
        LOWER(email_address) AS email,
        timestamp
    FROM task_1_response_2
),
mapped_responses AS (
    SELECT 
        m.student_id,
        r.email,
        r.timestamp
    FROM responses r
    LEFT JOIN student_email_map m
        ON LOWER(r.email) = LOWER(m.email)
),
ranked AS (
    SELECT 
        student_id,
        email,
        timestamp,
        ROW_NUMBER() OVER (
            PARTITION BY student_id 
            ORDER BY timestamp
        ) AS rn
    FROM mapped_responses
)

SELECT 
    student_id,
    email,
    timestamp,
    CASE 
        WHEN student_id IS NULL THEN 'Invalid'
        WHEN rn = 1 THEN 'Submitted'
        ELSE 'Duplicate'
    END AS status
FROM ranked;
```

```sql
-- Final Status (INCLUDING Not Submitted)
WITH student_base AS (
    SELECT 
        regno AS student_id,
        LOWER(COALESCE(college_domain_mailid, personal_mail_id)) AS email
    FROM training_enrolled
),
responses AS (
    SELECT 
        LOWER(email_address) AS email,
        timestamp
    FROM task_1_response_2
),
mapped_responses AS (
    SELECT 
        m.student_id,
        r.email,
        r.timestamp
    FROM responses r
    LEFT JOIN student_email_map m
        ON LOWER(r.email) = LOWER(m.email)
),
ranked AS (
    SELECT 
        student_id,
        ROW_NUMBER() OVER (
            PARTITION BY student_id 
            ORDER BY timestamp
        ) AS rn
    FROM mapped_responses
),
first_submission AS (
    SELECT student_id
    FROM ranked
    WHERE rn = 1
)

SELECT 
    s.student_id,
    CASE 
        WHEN f.student_id IS NOT NULL THEN 'Submitted'
        ELSE 'Not Submitted'
    END AS final_status
FROM student_base s
LEFT JOIN first_submission f
    ON s.student_id = f.student_id;
```

```sql
-- Count Submissions Per Student
SELECT 
    m.student_id,
    COUNT(*) AS submission_count
FROM task_1_response_2 r
LEFT JOIN student_email_map m
    ON LOWER(r.email_address) = LOWER(m.email)
WHERE m.student_id IS NOT NULL
GROUP BY m.student_id;
```

```sql
-- Students Using BOTH Emails
SELECT 
    m.student_id
FROM task_1_response_2 r
JOIN student_email_map m
    ON LOWER(r.email_address) = LOWER(m.email)
GROUP BY m.student_id
HAVING COUNT(DISTINCT LOWER(r.email_address)) > 1;
```

```sql
-- Summary Counts
WITH responses AS (
    SELECT 
        LOWER(email_address) AS email,
        timestamp
    FROM task_1_response_2
),
mapped_responses AS (
    SELECT 
        m.student_id,
        r.email,
        r.timestamp
    FROM responses r
    LEFT JOIN student_email_map m
        ON LOWER(r.email) = LOWER(m.email)
),
ranked AS (
    SELECT 
        student_id,
        ROW_NUMBER() OVER (
            PARTITION BY student_id 
            ORDER BY timestamp
        ) AS rn
    FROM mapped_responses
)

SELECT 
    CASE 
        WHEN student_id IS NULL THEN 'Invalid'
        WHEN rn = 1 THEN 'Submitted'
        ELSE 'Duplicate'
    END AS status,
    COUNT(*) AS count
FROM ranked
GROUP BY status;
```
